```python
import vrpSolver
```

```python
def readDat(fileName):
    dicNodeLoc = {}
    f = open(fileName, 'r')
    n = (int)(f.readline())
    for i in range(n):
        loc = f.readline()
        xy = loc.split("\t")
        dicNodeLoc[i] = ((float)(xy[0]), (float)(xy[1].split('\n')[0]))
    f.close()    
    return dicNodeLoc
```

```python
dicNodeLoc = readDat('TSP_instance_n_10_s_765.dat')
```

```python
res = vrpSolver.heuristic.consTSP(
    nodeLoc = dicNodeLoc,
    tau = 'Euclidean',
    algo = 'DepthFirst'
)
res
```

```python
vrpSolver.plotArcs(
    nodeLoc = dicNodeLoc,
    arcs = res['mst'],
    color='green'
)
```

```python
vrpSolver.plotSeq(
    nodeLoc = dicNodeLoc,
    seq = res['seq'],
    color='red'
)
```

```python
res = vrpSolver.heuristic.consTSP(
    nodeLoc = dicNodeLoc,
    tau = 'Euclidean',
    algo = 'Christofides'
)
res
```

```python
vrpSolver.plotArcs(
    nodeLoc = dicNodeLoc,
    arcs = res['mst'],
    color='green'
)
```

```python
vrpSolver.plotArcs(
    nodeLoc = dicNodeLoc,
    arcs = res['matching'],
    color='blue'
)
```

```python
vrpSolver.plotArcs(
    nodeLoc = dicNodeLoc,
    arcs = res['newGraph'],
    color='green'
)
```

```python
vrpSolver.plotSeq(
    nodeLoc = dicNodeLoc,
    seq = res['seq'],
    color='red'
)
```

```python
res = vrpSolver.exact.ipTSP(
    nodeLoc = dicNodeLoc,
    fml = 'DFJ_Lazy',
    tau = 'Euclidean',
    timeLimit = 100
)
res
```

```python
seq = [res['seq'][i] - 1 for i in range(len(res['seq']))]
vrpSolver.plotSeq(
    nodeLoc = dicNodeLoc,
    seq = seq
)
```
