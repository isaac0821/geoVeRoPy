```python
import os,sys,inspect
currentdir = os.getcwd()
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)

import vrpSolver
# %matplotlib notebook
```

```python
nodes = vrpSolver.rndNodes(
    nodeIDs = [i for i in range(1, 11)])
```

```python
radius = 10
nodes = vrpSolver.rndNodeNeighbors(
    nodes = nodes,
    shape = 'Circle',
    radius = radius)
```

```python
fig, ax = vrpSolver.plotNodes(
    nodes = nodes,
    nodeColor = 'black',
    boundingBox = (-20, 120, -20, 120))
for i in nodes:
    if ('neighbor' in nodes[i]):
        fig, ax = vrpSolver.plotPoly(
            fig = fig,
            ax = ax,
            edgeColor = 'black',
            poly = nodes[i]['neighbor'])
```

```python
cetsp = vrpSolver.solveCETSP(
    startLoc = (50, 50),
    endLoc = (50, 50),
    nodes = nodes,
    radius = radius,
    cutSetting = {
        'defaultCut': True
    })
```

```python
cetspMuilt = vrpSolver.solveCETSP(
    startLoc = (50, 50),
    endLoc = (50, 50),
    nodes = nodes,
    radius = radius,
    cutSetting = {
        'defaultCut': True,
        'multiCut': True
    })
```
