```python
f = open("../instance/bubbles3.ceop")

nodes = {}
startLoc = None
endLoc = None

a = f.readline()
l = a.split()
startLoc = (float(l[0]), float(l[1]))

a = f.readline()
l = a.split()
endLoc = (float(l[0]), float(l[1]))

while f:
    a = f.readline()
    a = a.replace("\n", "")
    l = a.split()
    nodes[len(nodes) + 1] = {
        'loc': (float(l[0]), float(l[1])),
        'radius': float(l[3]),
        'score': float(l[4])
    } 
```

```python
radius = 10

nodes = geoVeRoPy.rndNodes(
    nodeIDs = range(1, 21))
nodes = geoVeRoPy.rndNodeNeighbors(
    nodes = nodes,
    shape = 'Circle',
    radius = radius)
for n in nodes:
    nodes[n]['score'] = 100 * random.random()
```

```python
for n in nodes:
    nodes[n]['label'] = round(nodes[n]['score'], 2)
fig, ax = geoVeRoPy.plotNodes(
    nodes = nodes,
    boundingBox = (-20, 120, -20, 120))
fig, ax = geoVeRoPy.plotNodes(
    fig = fig,
    ax = ax,
    nodes = {'depot': {'loc': startLoc}},
    nodeMarkerSize = 3)
```

```python
ceop = geoVeRoPy.solveCEOP(
    startLoc = startLoc,
    endLoc = endLoc,
    nodes = nodes,
    popSize = 50,
    neighRatio = {
        'crossover': 1,
        'swap': 0.1,
        'exchange': 0.1,
        'rotate': 0.1
    },
    radius = 10,
    maxLength = 317.98 ,
    stop = {
        'numIter': 200
    }
)
```

```python
for p in ceop['popObj']:
    for n in nodes:
        nodes[n]['label'] = round(nodes[n]['score'], 2)
    fig, ax = geoVeRoPy.plotNodes(
        nodes = nodes,
        boundingBox = (-20, 120, -20, 120))
    fig, ax = geoVeRoPy.plotNodes(
        fig = fig,
        ax = ax,
        nodes = {'depot': {'loc': (100, 100)}},
        nodeMarkerSize = 3)
    fig, ax = geoVeRoPy.plotLocSeq(
        fig = fig,
        ax = ax,
        lineColor = 'red',
        locSeq = p.path)
    ax.set_title(f"Ofv: {p.score}; Dist: {p.dist}")
```

```python
for n in nodes:
    nodes[n]['label'] = round(nodes[n]['score'], 2)

visitedNodes = {}
unvisitedNodes = {}
for n in ceop['chromo'].turning:
    if (n != 0):
        visitedNodes[n] = nodes[n]
for n in ceop['chromo'].trespass:
    if (n != 0):
        visitedNodes[n] = nodes[n]
    
for n in nodes:
    if (n not in visitedNodes):
        unvisitedNodes[n] = nodes[n]
    
fig, ax = geoVeRoPy.plotNodes(
    nodes = visitedNodes,
    neighborColor = 'red',
    neighborOpacity = 0.3,
    boundingBox = (-20, 120, -20, 120))
if (len(unvisitedNodes) > 0):
    fig, ax = geoVeRoPy.plotNodes(
        fig = fig,
        ax = ax,
        nodes = unvisitedNodes,
        neighborColor = 'gray',
        neighborOpacity = 0.3,
        boundingBox = (-20, 120, -20, 120))
fig, ax = geoVeRoPy.plotNodes(
    fig = fig,
    ax = ax,
    nodes = {'depot': {'loc': (50, 50)}},
    nodeMarkerSize = 3)
fig, ax = geoVeRoPy.plotLocSeq(
    fig = fig,
    ax = ax,
    lineColor = 'red',
    locSeq = ceop['path'])
ax.set_title(f"Ofv: {ceop['score']}; Dist: {ceop['dist']}")
```
