```python
nodes = geoVeRoPy.rndNodes(
    nodeIDs = range(1, 6))
for n in nodes:
    nodes[n]['vec'] = geoVeRoPy.rndVec()
vehSpeed = 10
```

```python
startLoc = (50, 50)
endLoc = (50, 50)
mttsp = geoVeRoPy.solveMTTSP(
    startLoc = startLoc,
    endLoc = endLoc,
    nodes = nodes,
    vehSpeed = 10)
```

```python
startLoc = (50, 50)
endLoc = (50, 50)
mttsp = geoVeRoPyPri.solveMTTSP_MISOCP(
    startLoc = startLoc,
    endLoc = endLoc,
    nodes = nodes,
    vehSpeed = 10)
```

```python
maxTime = mttsp['timedSeq'][-1][1] + 5
for n in nodes:
    (spd, deg) = geoVeRoPy.vecXY2Polar(nodes[n]['vec'])
    dist = maxTime * spd
    finalLoc = geoVeRoPy.ptInDistXY(nodes[n]['loc'], deg, dist)
    nodes[n]['timedSeq'] = [(nodes[n]['loc'], 0), (finalLoc, maxTime)]
```

```python
vehicles = {
    'veh': {'timedSeq': mttsp['timedSeq']}
}
```
