```python
import os,sys,inspect
currentdir = os.getcwd()
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)

import vrpSolver
```

```python
def readInstanceCVRPTW(fileName):
    f = open(fileName, 'r')
    
    line = f.readline()
    fields = str.split(line)
    instanceName = fields[0]
    vehNum = int(fields[1])
    vehCap = int(fields[2])
    
    nodes = {}
    for line in f:
        fields = line.split()
        try:
            nodes[int(fields[0])] = {
                'loc': (float(fields[1]), float(fields[2])),
                'demand': float(fields[3]),
                'timeWindow': [float(fields[4]), float(fields[5])],
                'serviceTime': float(fields[6])
            }
        except:
            pass
    
    return instanceName, vehNum, vehCap, nodes
```

```python
instanceName, numVeh, vehCap, nodes = readInstanceCVRPTW("../instance/C101.txt")
```

```python
nodes
```

```python
vrptw = vrpSolver.solveVRPTW(
    nodes = nodes,
    numVeh = 100,
    vehCap = vehCap,
    depotID = 1,
    customerIDs = 'AllButDepot'
)
```
